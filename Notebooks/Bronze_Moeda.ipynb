```python
import requests as rq
```

```python
url = 'https://olinda.bcb.gov.br/olinda/servico/PTAX/versao/v1/odata/Moedas?$top=100&$format=json&$select=simbolo,nomeFormatado'
response = rq.get(url)
dados = response.json()['value']
moedas = spark.createDataFrame(dados)
# display(moedas)

moedas = moedas.selectExpr(
    "nomeFormatado AS NomeMoeda",
    "Simbolo AS Moeda"
)

moedas.write.mode('overwrite').saveAsTable('Moedas')
```
